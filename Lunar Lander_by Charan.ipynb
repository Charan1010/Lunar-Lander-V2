```python
gamma = 0.95
```

```python
def discount_rewards(rewards,y_episode):
        discounted_rewards = [0]*len(rewards)
        running_add = 0
        for t in reversed(range(0, len(rewards))):
            running_add = running_add * gamma + rewards[t]
            discounted_rewards[t] = running_add
            
        discounted_rewards -= np.mean(discounted_rewards)
        discounted_rewards /= np.std(discounted_rewards)
        #print discounted_rewards
        for i in range(len(y_episode)):
            y_episode[i]=y_episode[i]*discounted_rewards[i]
            
        return y_episode
```

```python
def gather_data(env,model):
    episodes=3000
    steps=300
    epsilon=0.3
    min_score=50
    x_train,y_train=[],[]
    scores=[]
    #count=0
    for episode in range(episodes):
        print "Running episode "+str(episode)
        observation=env.reset()
        score=0
        score_vector=[]
        x_episode,y_episode=[],[]
        for step in range(steps):
            #print "In Step"+str(step)
            if random.uniform(0, 1)>epsilon:
                action=np.random.randint(0,4)
            #else:
            action=np.argmax(model.predict(observation.reshape(1,8)))
            one_hot_action=np.zeros(4)
            one_hot_action[action]=1
            x_episode.append(observation)
            y_episode.append(one_hot_action)
            observation,reward,done,_=env.step(action)
            #env.render()
            score+=reward
            score_vector.append(reward)
            if done:
                break
        if score>min_score:  
            x_train+=x_episode
            y_train+=discount_rewards(score_vector,y_episode)
            scores.append(score)
        #if count>100:
        #    np.save("lander_x.npy",np.array(x_train))
        #    np.save("lander_y.npy",np.array(y_train))
        #    count=0
        #count+=1
        
    x_train=np.array(x_train)
    y_train=np.array(y_train)
    
    return x_train,y_train
    
            
                
    
```

```python
def get_model():
    model = Sequential()
    model.add(Dense(128, input_shape=(8,), activation="relu"))
    model.add(Dropout(0.4))

    model.add(Dense(256, activation="relu"))
    model.add(Dropout(0.4))

    #model.add(Dense(512, activation="relu"))
    #model.add(Dropout(0.4))

    #model.add(Dense(256, activation="relu"))
    #model.add(Dropout(0.4))

    model.add(Dense(128, activation="relu"))
    model.add(Dropout(0.4))
    model.add(Dense(4, activation="softmax"))

    model.compile(
        loss="categorical_crossentropy",
        optimizer="rmsprop",
        metrics=["accuracy"])
    return model
```

```python
def main():
    env = gym.make("LunarLander-v2")
    
   
    try:
        print "model already exists"
        model=keras.models.load_model('lander_basic3.h5')
    except:
        print "Model isnt created yet"
        model=get_model()
    #model=get_model()
    x_train,y_train=gather_data(env,model)
    
    print "Ended collection of data"
    model.fit(x_train,y_train,epochs=5)
    model.save('lander_basic3.h5')
    print "model saved"
    
    scores=[]
    episodes=50
    steps=500
    print "STARTED TESTING"
    for episode in range(episodes):
        print "Testing episode"+str(episode)
        observation=env.reset()
        score=0
        for step in range(steps):
            print "In step"+str(step)
            action=np.argmax(model.predict(observation.reshape(1,8)))
            observation,reward,done,_=env.step(action)
            #env.render()
            score+=reward
            if done:
                break
        scores.append(score)
    print "The average score is "+str(np.mean(scores))
    print scores
    return [np.mean(scores),scores]
        
    
```
